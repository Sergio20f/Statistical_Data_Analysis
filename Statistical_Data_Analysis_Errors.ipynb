```python
"""
Once again we use the iris data from sklearn.  We need matplotlib for plotting out scatter plots and histograms,
and we will be using numpy to compute covaraince and correlation matrices. So all three are imported at the
start of this notebook.
"""
print("Import the sklearn datasets")
import sklearn.datasets as ds
import numpy as np
from scipy import stats


"""
The Iris data is to be used for the study.
"""

print("Load the iris data")
Bunch = ds.load_iris(return_X_y=False) # we want to get the Bunch so that we can access more than just the data and target values
target_name = "class"
target_plot_xaxis = "target value (class)"
print("The iris data sets have been loaded")
```

Now we can proceed to compute the mean and uncertainty for each of the types of Iris data.  This will allow us to compare each of the four variables, sepal width, sepal length, petal width and petal length for each type of iris to determine if it is possible to distinguish between one or more types of iris using these quantities alone. 

```python
sepal_length_sentosa = []
sepal_length_versicolor = []
sepal_length_virginica = []

sepal_width_sentosa = []
sepal_width_versicolor = []
sepal_width_virginica = []

petal_length_sentosa = []
petal_length_versicolor = []
petal_length_virginica = []

petal_width_sentosa = []
petal_width_versicolor = []
petal_width_virginica = []

for idx, data in enumerate(Bunch['data']):
    if Bunch['target'][idx] == 0:
        sepal_length_sentosa.append(data[0])
        sepal_width_sentosa.append(data[1])
        petal_length_sentosa.append(data[2])
        petal_width_sentosa.append(data[3])
        
    elif Bunch['target'][idx] == 1:
        sepal_length_versicolor.append(data[0])
        sepal_width_versicolor.append(data[1])
        petal_length_versicolor.append(data[2])
        petal_width_versicolor.append(data[3])
    
    else:
        sepal_length_virginica.append(data[0])
        sepal_width_virginica.append(data[1])
        petal_length_virginica.append(data[2])
        petal_width_virginica.append(data[3])

print("For Sentosa iris: ")
print("Sepal length mean (cm): " + str(np.mean(sepal_length_sentosa)))
print("Sepal length standard deviation (cm): " + str(np.std(sepal_length_sentosa)))
print("\n")
print("Sepal width mean (cm): " + str(np.mean(sepal_width_sentosa)))
print("Sepal width standard deviation (cm): " + str(np.std(sepal_width_sentosa)))
print("\n")
print("Petal length mean (cm): " + str(np.mean(petal_length_sentosa)))
print("Petal length standard deviation (cm): " + str(np.std(petal_length_sentosa)))
print('\n')
print("Petal width mean (cm): " + str(np.mean(petal_width_sentosa)))
print("Petal width standard deviation (cm): " + str(np.std(petal_width_sentosa)))

print("\n")

print("For Versicolor iris: ")
print("Sepal length mean (cm): " + str(np.mean(sepal_length_versicolor)))
print("Sepal length standard deviation (cm): " + str(np.std(sepal_length_versicolor)))
print("\n")
print("Sepal width mean (cm): " + str(np.mean(sepal_width_versicolor)))
print("Sepal width standard deviation (cm): " + str(np.std(sepal_width_versicolor)))
print("\n")
print("Petal length mean (cm): " + str(np.mean(petal_length_versicolor)))
print("Petal length standard deviation (cm): " + str(np.std(petal_length_versicolor)))
print('\n')
print("Petal width mean (cm): " + str(np.mean(petal_width_versicolor)))
print("Petal width standard deviation (cm): " + str(np.std(petal_width_versicolor)))

print("\n")

print("For Virginica iris: ")
print("Sepal length mean (cm): " + str(np.mean(sepal_length_virginica)))
print("Sepal length standard deviation (cm): " + str(np.std(sepal_length_virginica)))
print("\n")
print("Sepal width mean (cm): " + str(np.mean(sepal_width_versicolor)))
print("Sepal width standard deviation (cm): " + str(np.std(sepal_width_virginica)))
print("\n")
print("Petal length mean (cm): " + str(np.mean(petal_length_virginica)))
print("Petal length standard deviation (cm): " + str(np.std(petal_length_virginica)))
print('\n')
print("Petal width mean (cm): " + str(np.mean(petal_width_virginica)))
print("Petal width standard deviation (cm): " + str(np.std(petal_width_virginica)))
```

```python
def comparator(mean_1, mean_2, std_1, std_2, var_1=str, var_2=str):
    diff = abs(mean_1 - mean_2)
    quad_sum = np.sqrt(std_1**2 + std_2**2)
    sigma_test = diff/quad_sum
    
    if sigma_test > 3:
        return "Can distinguish between " + var_1 + " and " + var_2 
    else:
        return "Cannot distinguish between " + var_1 + " and " + var_2

concl_sepal_length_1 = comparator(np.mean(sepal_length_sentosa), np.mean(sepal_length_versicolor),
                     np.std(sepal_length_sentosa), np.std(sepal_length_versicolor),
                    "sepal length sentosa", "sepal length versicolor")

concl_sepal_length_2 = comparator(np.mean(sepal_length_sentosa), np.mean(sepal_length_virginica),
                    np.std(sepal_length_sentosa), np.std(sepal_length_virginica),
                    "sepal length sentosa", "sepal length virginica")

concl_sepal_length_3 = comparator(np.mean(sepal_length_virginica), np.mean(sepal_length_versicolor),
                     np.std(sepal_length_virginica), np.std(sepal_length_versicolor),
                    "sepal length virginica", "sepal length versicolor")

concl_sepal_width_1 = comparator(np.mean(sepal_width_sentosa), np.mean(sepal_width_versicolor),
                                np.std(sepal_width_sentosa), np.std(sepal_width_versicolor),
                                "sepal width sentosa", "sepal width versicolor")

concl_sepal_width_2 = comparator(np.mean(sepal_width_sentosa), np.mean(sepal_width_virginica),
                                np.std(sepal_width_sentosa), np.std(sepal_width_virginica),
                                "sepal width sentosa", "sepal width virginica")

concl_sepal_width_3 = comparator(np.mean(sepal_width_virginica), np.mean(sepal_width_versicolor),
                                np.std(sepal_length_virginica), np.std(sepal_width_versicolor),
                                "sepal width virginica", "sepal width versicolor")

concl_petal_length_1 = comparator(np.mean(petal_length_sentosa), np.mean(petal_length_versicolor),
                                 np.std(petal_length_sentosa), np.std(petal_length_versicolor),
                                 "petal length sentosa", "petal length versicolor")

concl_petal_length_2 = comparator(np.mean(petal_length_sentosa), np.mean(petal_length_virginica),
                                 np.std(petal_length_sentosa), np.std(petal_length_virginica),
                                 "petal length sentosa", "petal length virginica")

concl_petal_length_3 = comparator(np.mean(petal_length_versicolor), np.mean(petal_length_virginica),
                                 np.std(petal_length_versicolor), np.std(petal_length_virginica),
                                 "petal length versicolor", "petal length virginica")

concl_petal_width_1 = comparator(np.mean(petal_width_sentosa), np.mean(petal_width_versicolor),
                               np.std(petal_width_sentosa), np.std(petal_width_versicolor),
                               "petal width sentosa", "petal width versicolor")

concl_petal_width_2 = comparator(np.mean(petal_width_sentosa), np.mean(petal_width_virginica),
                                np.std(petal_width_sentosa), np.std(petal_width_virginica),
                                "petal width sentosa", "petal width virginica")

concl_petal_width_3 = comparator(np.mean(petal_width_versicolor), np.mean(petal_width_virginica),
                                np.std(petal_width_versicolor), np.std(petal_width_virginica),
                                "petal width versicolor", "petal width virginica")

print(concl_sepal_length_1)
print(concl_sepal_length_2)
print(concl_sepal_length_3)

print("\n")

print(concl_sepal_width_1)
print(concl_sepal_width_2)
print(concl_sepal_width_3)

print("\n")

print(concl_petal_length_1)
print(concl_petal_length_2)
print(concl_petal_length_3)

print("\n")

print(concl_petal_width_1)
print(concl_petal_width_2)
print(concl_petal_width_3)
```
